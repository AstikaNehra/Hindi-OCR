```python
import cv2
import re
import os
import numpy as np
import pandas as pd

def writefile(data,path):
    df=pd.DataFrame(np.array(data,dtype="object"))
    with open(path,'a+') as f:
        df.to_csv(f,mode='a',header=False)

def load_images_from_folder(folder,letter,output,ctr):
    images = []
    for filename in os.listdir(folder):
        ctr+=1
        final=[]
        img = cv2.imread(os.path.join(folder,filename),0)
        img=255-img
        final.append([str(letter)]+list(img.flatten()))
        writefile(final,output)
        if img is not None:
            images.append(img)
    return ctr

# path='nhcd/numerals/'
# ctr=0
# for i in range(0,10):
#         ctr=load_images_from_folder(path+str(i),str(i),'numerals.csv',ctr)
#         print(i)
# print("Numerals are",ctr)
# ctr=0
# consonants=['ka','kha','ga','gha','kna','cha','chha','ja','jha','yna','ta','tha','da','dha','ana','taa','thaa','daa','dhaa','na','pa','pha','ba','bha','ma','ya','ra','la','va','motosaw','petchiryosaw','patalosaw','ha','ksha','tra','gya']
# path='nhcd/consonants/'
# for i in range(0,36):
#         ctr=load_images_from_folder(path+str(i+1),consonants[i],'consonants.csv',ctr)
#         print(consonants[i])
# print("Consonants are",ctr)
ctr=0
path='nhcd/vowels/'
vowels=['a','aa','i','ee','u','oo','ae','ai','o','au','an','ah']
for j in range(0,30):
        for i in range(0,10):
                ctr=load_images_from_folder(path+str(i+1),vowels[i],'vowels.csv',ctr)
                print(j,vowels[i])
print("Vowels are",ctr)
print("Dataset ready!")
```
